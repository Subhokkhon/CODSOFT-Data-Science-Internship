```python
column_lst = ["gender", "age", "annual_salary", "credit_card_debt", "net_worth", "car_purchase_amount"]
for i in range(len(column_lst[:-1])):
    ax = plt.subplots(figsize=(5, 5))
    sns.regplot(data=data, x=column_lst[i], y=column_lst[-1])
    plt.xticks(rotation=40)
    plt.title(column_lst[i].upper())
```

```python
data.corr()
```

```python
x = data[column_lst[:-1]]
y = data['car_purchase_amount']
train_x, val_x, train_y, val_y = train_test_split(x, y, test_size=0.1, random_state=4)
```

```python
lm = LinearRegression()
lm.fit(train_x, train_y)

m1 = lm.predict(val_x)
m1
```

```python
rf = RandomForestRegressor()
rf.fit(train_x, train_y)
m2 = rf.predict(val_x)
m2
```

```python
dt = DecisionTreeRegressor()
dt.fit(train_x, train_y)
m3 = dt.predict(val_x)
m3
```

```python
ls = Lasso()
ls.fit(train_x, train_y)
m4 = ls.predict(val_x)
m4
```

```python
rd = Ridge()
rd.fit(train_x, train_y)
m5 = rd.predict(val_x)
m5
```

```python
model_lst = [m1, m2, m3, m4, m5]
model_names = ["Linear Regression", "Random Forest", "Decision Tree", "Lasso", "Ridge"]
for i in range(len(model_names)):
    ax = plt.subplots(figsize=(5, 5))
    sns.regplot(x=model_lst[i], y=val_y)
    plt.xticks(rotation=40)
    plt.title(model_names[i].upper() + " MODEL")
    plt.xlabel("Predicted Values")
    plt.ylabel("Actual Values")
```

```python
for i in range(len(model_names)):
    print(f"Mean absolute error of {model_names[i]} = {mse(val_y, model_lst[i])}")
lm_score = r2_score(val_y, m1)
rf_score = r2_score(val_y, m2)
dt_score = r2_score(val_y, m3)
ls_score = r2_score(val_y, m4)
rd_score = r2_score(val_y, m5)

accuracy = {
    "Multiple Linear Regression": lm_score,
    "Random Forest Regressor": rf_score,
    "Decision Tree Regressor": dt_score,
    "Lasso": ls_score,
    "Ridge": rd_score,
}
# accuracy
print(f"\nThe {max(accuracy, key=accuracy.get)} model has the highest accuracy with value {max(accuracy.values()) * 100}%")
```
